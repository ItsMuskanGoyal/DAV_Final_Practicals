```python
import numpy as np
arr1 = np.random.rand(4,4)
arr2 = np.random.rand(4,4)
arr3 = np.random.rand(4,4)
```

```python
print("Original array1:")
print(arr1)
print("Original array2:")
print(arr2)
print("Original array3:")
print(arr3)
```

```python
arr4 = arr3-arr2
print("Array4:")
print(arr4)
```

```python
arr5 = arr1*2
print("Array5:")
print(arr5)
```

```python

print("Covariance of Array1 and Array4:")
print(np.cov(arr1,arr4))
print("Correlation of Array1 and Array5:")
print(np.corrcoef(arr1,arr5))
```
